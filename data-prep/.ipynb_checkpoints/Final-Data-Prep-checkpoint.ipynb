```python
import pandas as pd
import datetime 
from pandas.tseries.offsets import MonthEnd
```

```python
df_cases = pd.read_csv\
    ('./MD_COVID-19_-_Cases_by_Age_Distribution.csv')
df_cases['Totals'] = df_cases.iloc[:,2:27].sum(axis=1)
df_cases['DATE'] = pd.to_datetime(df_cases['DATE'])
# df_cases['MONTH'] = pd.to_datetime(df_cases['DATE']).dt.strftime('%Y-%B')
df_cases = df_cases.fillna(0)
df_cases = df_cases.drop(columns=['OBJECTID'])
df_cases = df_cases.melt(id_vars='DATE',var_name='AGE',value_name='CASES')\
    .groupby(['DATE','AGE'])['CASES'].sum().reset_index()

df_cases['CASE_CNT'] = df_cases.groupby(['AGE'])['CASES'].diff().fillna(0)

df_cases['DATE'] = pd.to_datetime(df_cases['DATE'], format="%Y%m") + MonthEnd(1)




df_cases = df_cases[(df_cases['DATE'] >= '2020-1-1') &\
                    (df_cases['DATE'] <= '2021-7-1')]

df_cases['MONTH'] = df_cases['DATE'].dt.strftime("%B %Y")

df_cases = df_cases.groupby(['MONTH','DATE','AGE'])['CASE_CNT'].sum().reset_index()


df_cases['AGE'] = df_cases['AGE'].str.replace('Age_','')
df_cases['AGE'] = df_cases['AGE'].str.replace('_to_','-')
df_cases['AGE'] = df_cases['AGE'].str.replace('plus','+')
df_cases['DATE'] = df_cases['DATE'].dt.strftime('%Y-%m-%d')

df_cases = df_cases.sort_values(by=['AGE','DATE']).reset_index()

df_cases = df_cases.drop(columns=['index'])

#df_cases['CASE_PER'] = \
#((df_cases.set_index(['MONTH','DATE'])['CASE_CNT']*100\
#    /df_cases[~df_cases['AGE'].str.contains('Totals')]\
#      .groupby(['MONTH','DATE'])['CASE_CNT'].sum()))
#df_cases[df_cases['MONTH'].str.contains('January')]


df_totals = df_cases[~df_cases['AGE'].str.contains('Totals')]\
    .groupby(['MONTH','DATE'])['CASE_CNT'].sum().reset_index()\
    .rename(columns={"CASE_CNT":"CASE_TOTAL"})


df_cases = df_cases.merge(df_totals,how = 'left', on=['MONTH','DATE'])
df_cases['CASE_PER'] = df_cases['CASE_CNT'] *100 /df_cases['CASE_TOTAL']
df_cases[df_cases['MONTH'].str.contains('Mar')]
df_cases=df_cases.drop(columns=['CASE_TOTAL'])

df_cases
```

```python
df_deaths = pd.read_csv\
    ('./MD_COVID-19_-_Confirmed_Deaths_by_Age_Distribution.csv')
df_deaths['Totals'] = df_deaths.iloc[:,2:27].sum(axis=1)
df_deaths['DATE'] = pd.to_datetime(df_deaths['DATE'])
df_deaths = df_deaths.fillna(0)
df_deaths = df_deaths.drop(columns=['OBJECTID'])

df_deaths['Age_Unknown'] =df_deaths['Age_Unknown'].cumsum()

df_deaths = df_deaths.melt(id_vars='DATE',var_name='AGE',value_name='DEATHS')\
    .groupby(['DATE','AGE'])['DEATHS'].sum().reset_index()

#df_deaths = df_deaths[df_deaths['AGE'].str.contains('Totals')]\
#    .set_index(['AGE','DATE']).diff().fillna(0).reset_index()

df_deaths['DEATH_CNT'] = df_deaths.groupby(['AGE'])['DEATHS'].diff().fillna(0)

df_deaths['DATE'] = pd.to_datetime(df_deaths['DATE'], format="%Y%m") + MonthEnd(1)


df_deaths = df_deaths[(df_deaths['DATE'] >= '2020-1-1') &\
                    (df_deaths['DATE'] <= '2021-7-1')]

df_deaths['MONTH'] = df_deaths['DATE'].dt.strftime("%B %Y")
df_deaths = df_deaths.groupby(['MONTH','DATE','AGE'])['DEATH_CNT'].sum().reset_index()

df_deaths['AGE'] = df_deaths['AGE'].str.replace('Age_','')
df_deaths['AGE'] = df_deaths['AGE'].str.replace('_to_','-')
df_deaths['AGE'] = df_deaths['AGE'].str.replace('plus','+')

df_deaths['DATE'] = df_deaths['DATE'].dt.strftime('%Y-%m-%d')

df_deaths = df_deaths.sort_values(by=['AGE','DATE']).reset_index()

df_deaths = df_deaths.drop(columns=['index'])

df_totals = df_deaths[~df_deaths['AGE'].str.contains('Totals')]\
    .groupby(['MONTH','DATE'])['DEATH_CNT'].sum().reset_index()\
    .rename(columns={"DEATH_CNT":"DEATH_TOTAL"})


df_deaths = df_deaths.merge(df_totals,how = 'left', on=['MONTH','DATE'])
df_deaths['DEATH_PER'] = df_deaths['DEATH_CNT'] *100 /df_deaths['DEATH_TOTAL']
df_deaths[df_deaths['MONTH'].str.contains('Mar')]
df_deaths=df_deaths.drop(columns=['DEATH_TOTAL'])

df_deaths
```

```python
df_vac = pd.read_csv\
    ('./MD_COVID-19_-_Vaccinations_by_Age_Distribution.csv')
df_vac = df_vac[df_vac['AgeRange'].notnull()].fillna(0)
df_vac = df_vac[~df_vac['AgeRange'].str.contains('Federal Entities')]

df_vac['DATE'] = pd.to_datetime(df_vac['VACCINATION_DATE'])

df_vac = df_vac[['DATE','AgeRange','CombinedTotalDosesCumulative']]
df_vac['VAC_CNT'] = df_vac.groupby(['AgeRange'])['CombinedTotalDosesCumulative']\
    .diff().fillna(0)

df_vac = df_vac.rename(columns={"AgeRange":"AGE"})

df_vac['DATE'] = pd.to_datetime(df_vac['DATE'], format="%Y%m") + MonthEnd(1)




df_vac = df_vac[(df_vac['DATE'] >= '2020-1-1') &\
                    (df_vac['DATE'] <= '2021-7-1')]

df_vac['MONTH'] = df_vac['DATE'].dt.strftime("%B %Y")

df_vac = df_vac.groupby(['MONTH','DATE','AGE'])['VAC_CNT'].sum().reset_index()

df_vac['AGE'] = df_vac['AGE'].str.replace('age','')
df_vac['AGE'] = df_vac['AGE'].str.replace('to','-')
df_vac['AGE'] = df_vac['AGE'].str.replace('plus','+')


df_temp = df_vac.groupby(['DATE'])['VAC_CNT'].sum().reset_index() #df_vac
df_temp['AGE'] = 'Totals'
df_temp['MONTH'] = pd.to_datetime(df_temp['DATE']).dt.strftime("%B %Y")
df_temp

df_vac['DATE'] = df_vac['DATE'].dt.strftime('%Y-%m-%d')
df_temp['DATE'] = df_temp['DATE'].dt.strftime('%Y-%m-%d')
df_vac = df_vac.append(df_temp)
#df_vac.to_csv(r'./project/dataset/Vac_by_Age.csv_22',index = False)

df_vac = df_vac.sort_values(by=['AGE','DATE']).reset_index()
df_vac = df_vac.drop(columns=['index'])


df_totals = df_vac[~df_vac['AGE'].str.contains('Totals')]\
    .groupby(['MONTH','DATE'])['VAC_CNT'].sum().reset_index()\
    .rename(columns={"VAC_CNT":"VAC_TOTAL"})


df_vac = df_vac.merge(df_totals,how = 'left', on=['MONTH','DATE'])
df_vac['VAC_PER'] = df_vac['VAC_CNT'] *100 /df_vac['VAC_TOTAL']
df_vac[df_vac['MONTH'].str.contains('Mar')]
df_vac=df_vac.drop(columns=['VAC_TOTAL'])

df_vac
```
